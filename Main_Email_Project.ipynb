```python
import time
import imaplib
import email
from email.header import decode_header

# Your Gmail Credentials
EMAIL_ACCOUNT = "sanket8412@gmail.com"
EMAIL_PASSWORD = "uncb zgcq mioo nhof"

# Set to store already processed emails
processed_emails = set()

def fetch_all_emails():
    """Fetch all emails from the first email to the latest and monitor for new ones."""
    while True:
        try:
            # Connect to Gmail
            mail = imaplib.IMAP4_SSL("imap.gmail.com")
            mail.login(EMAIL_ACCOUNT, EMAIL_PASSWORD)
            mail.select("inbox")

            # Search for all emails from the specific sender
            status, messages = mail.search(None, 'FROM "tukaram.tanpure@planeteyefarm.ai"')

            # Get all email IDs
            email_ids = messages[0].split()

            if not email_ids:
                print("‚ùå No emails found.")
            else:
                print(f"üì¨ Found {len(email_ids)} emails!")

                # Process emails from oldest to newest
                for email_id in email_ids:
                    if email_id in processed_emails:
                        continue  # Skip already processed emails

                    # Fetch the email
                    status, msg_data = mail.fetch(email_id, "(RFC822)")
                    raw_email = msg_data[0][1]
                    msg = email.message_from_bytes(raw_email)

                    # Decode subject
                    subject, encoding = decode_header(msg["Subject"])[0]
                    if isinstance(subject, bytes):
                        subject = subject.decode(encoding if encoding else "utf-8")

                    # Print email details
                    print(f"\nüì© New Email Received from {msg['From']}")
                    print(f"üìú Subject: {subject}")

                    # Extract and print email body
                    if msg.is_multipart():
                        for part in msg.walk():
                            content_type = part.get_content_type()
                            content_disposition = str(part.get("Content-Disposition"))

                            if content_type == "text/plain" and "attachment" not in content_disposition:
                                body = part.get_payload(decode=True).decode()
                                print("\nüìÑ Email Body:\n", body)
                    else:
                        body = msg.get_payload(decode=True).decode()
                        print("\nüìÑ Email Body:\n", body)

                    # Mark email as processed
                    processed_emails.add(email_id)

            # Logout and wait before checking again
            mail.logout()

        except Exception as e:
            print(f"‚ö†Ô∏è Error: {e}")

        print("üîÑ Checking again in 30 seconds...")
        time.sleep(30)  # Adjust this time as needed

# Start fetching all emails and monitor new ones
fetch_all_emails()
```
